```python
import matplotlib.pyplot as plt
# Generating the dataset
people = ['Kiran', 'Arun', 'Vijay', 'Varun']
age = [25, 30, 35, 40]
height = [145, 151, 165, 173]
weight = [45, 55, 65, 75]

# Scatter Plot
plt.figure(figsize=(8, 6))
plt.scatter(age, height, s=weight, alpha=0.7, c='blue')
plt.title('Relationship between Age, Height, and Weight')
plt.xlabel('Age')
plt.ylabel('Height')
plt.legend(['Weight'], loc='upper left')
plt.grid(True)
plt.show()

# Bar Chart
plt.figure(figsize=(8, 6))
plt.bar(people, height, color='green')
plt.title('Height of People')
plt.xlabel('People')
plt.ylabel('Height')
plt.grid(axis='y')
plt.show()

# Histogram
plt.figure(figsize=(8, 6))
plt.hist(weight, bins=3, color='orange', edgecolor='black')
plt.title('Weight Distribution')
plt.xlabel('Weight')
plt.ylabel('Frequency')
plt.grid(axis='y')
plt.show()
```
